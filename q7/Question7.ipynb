```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
film_df = pd.read_csv("../Resources/film_df.csv")
film_df
```

```python
#Budget and worldwide_gross vs time
money_time = film_df[['date','title','budget','worldwide_gross']].sort_values(by="date")
money_time.set_index("date")

plt.bar(money_time['title'],money_time['budget'])
plt.xticks(rotation=90)
plt.ylim(0,400000000)

plt.title('Marvel Budget By Title')
plt.xlabel("Movies")
plt.ylabel("Budget (Tens of Millions)")
plt.grid(True)
```

```python
#Budget and worldwide_gross vs time
money_time = film_df[['date','title','budget','worldwide_gross']].sort_values(by="date")
money_time.set_index("date")

plt.bar(money_time['title'],money_time['worldwide_gross'])
plt.xticks(rotation=90)
plt.ylim(0,3000000000)

plt.title('Marvel Worldwide Gross By Title')
plt.xlabel("Movies")
plt.ylabel("Worldwide Gross (Billions)")
plt.grid(True)
```

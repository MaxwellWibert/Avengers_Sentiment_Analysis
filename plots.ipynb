```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
```

```python
film_df = pd.read_csv("./Resources/film_df.csv").set_index('title')
```

```python
film_df
```

```python
def magnify():
    return [dict(selector="th",
                 props=[("font-size", "5pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')]),
            dict(selector="th.col_heading",
                 props=[("writing-mode", "vertical-rl"), 
                        ('transform', 'rotateZ(-90deg)'),
                        ('padding', '4em 0em')
                        ])
           ]
```

```python
# correlation matrix
corr = film_df.corr()
corr.style.background_gradient(cmap='coolwarm', axis =1).set_properties(**{'max-width': '80px', 'font-size': '1pt'}).set_table_styles(magnify())

# plt.matshow(film_df.corr())
# plt.show()
```
